```python
print('='*80)
print('*'*32,"ELECTIONS 2022",'*'*32)
print('='*80)
print('*'*32,"List of candidates",'*'*32,)
print('='*80)
print("Enter M for Modi")
print("Enter R for Rahul")
print("Enter Y for Yogi")
print("Enter B for Biden")
print("Enter T for Trump")
print("Enter K for Kejriwal")
print("Enter G for GTM")
print("Enter N for NOTA")
print("Enter Q for Quit")
print('='*80)
Modi=0
Rahul=0
Yogi=0
Biden=0
Trump=0
Kejriwal=0
GTM=0
NOTA=0
while True:
    Ent=input("Enter your vote for Candidates: ")
    if Ent.upper()=='M':
        Modi+=1
    elif Ent.upper()=='R':
        Rahul+=1
    elif Ent.upper()=='Y':
        Yogi+=1
    elif Ent.upper()=='B':
        Biden+=1
    elif Ent.upper()=='T':
        Trump+=1
    elif Ent.upper()=='K':
        Kejriwal+=1
    elif Ent.upper()=='G':
        GTM+=1
    elif Ent.upper()=='Q':
        break
    else:
        NOTA+=1

print('*'*80)        
print("Modi has got ",Modi," votes")
print("Rahul has got ",Rahul," votes")
print("Yogi has got ",Yogi," votes")
print("Biden has got ",Biden," votes")
print("Trump has got ",Trump," votes")
print("Kejriwal has got ",Kejriwal," votes")
print("GTM has got ",GTM," votes")
print("People who have choosen None Of The Above are: ",NOTA)
print('*'*80)

dict1={Modi:'Modi',Rahul:'Rahul',Yogi:'Yogi',Biden:'Biden',Trump:'Trump',Kejriwal:'Kejriwal',GTM:'GTM',NOTA:'NOTA'}
List1=list(dict1.keys())
a=max(List1)
win=dict1.get(a)
print('='*80)
print('*'*23,"Winner of the Election is ",win,'*'*23)
print('='*80)
```
