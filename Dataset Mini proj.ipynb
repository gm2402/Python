```python
print('*'*100)
lis1=[1,2,3,'Gtm','RJ',2.66,3.44,'23']
print('lis1=',lis1)
print('='*100)
print("Enter 1 for creating other list")
print("Enter 2 to make changes in list")
print("Enter 3 for add elements in list")
print("Enter 4 for delete element from list ")
print('='*100)
a=(int(input("Enter choices for List: ")))
if a==1:
    print("-"*100)
    print("Enter 1 for normal list ")
    print("Enter 2 for empty list")
    print("-"*100)
    b=(int(input("Enter your choice for list creation ")))
    if(b==1):
        lis1=(list(eval(input("Enter list elements "))))
        print("You just created a list name lis1 ")
        print('lis1=',lis1)
    elif(b==2):
        lis1=[]
        print("You just created an empty list named lis3 ")
        print("lis1= ",lis1)
    else:
        print("Invalid Choice")
elif a==2:
    print("Enter 1 for making changes through indexing ")
    print("Enter 2 for making changes through slicing ")
    c=(int(input("Enter method for changes in your list ")))
    if(c==1):
        ind0=(int(input("Enter index for updating ")))
        ind1=(eval(input("Enter element you want to add ")))
        lis1[ind0]=ind1
        print(lis1)
    elif(c==2):
        slic0=(int(input("Enter start index position ")))
        slic1=(int(input("Enter stop index position ")))
        slic2=(int(input("Enter steps for slicing ")))
        slic3=(eval(input("Enter elements for updating list ")))
        lis1[slic0:slic1:slic2]=slic3
        print(lis1)
elif a==3:
    print("Enter 1 for adding single element at end of the list ")
    print("Enter 2 for adding single element at specified position ")
    print("Enter 3 for adding multiple element at end of the list ")
    d=(int(input("Enter method for adding elements in list ")))
    if(d==1):
        app1=(eval(input("Enter an element to add at end of the list ")))
        lis1.append(app1)
        print(lis1)
    elif(d==2):
        ins1=(int(input("Enter specific position to add in the list ")))
        ins2=(eval(input("Enter element to add in the list ")))
        lis1.insert(ins1,ins2)
        print(lis1)
    elif(d==3):
        ext1=(eval(input("Enter multiple elements to add in the list ")))
        lis1.extend(ext1)
        print(lis1)
    else:
        print("Invalid Choice")
elif a==4:
    print("Enter 1 for using pop() method ")
    print("Enter 2 for using remove() method ")
    print("Enter 3 for using clear() method ")
    print("Enter 4 for using delete() method ")
    e=(int(input("Enter choice for deletion method ")))
    if(e==1):
        print("Enter 1 for using pop() first method ")
        print("Enter 2 for using pop('Index') second method ")
        f=(int(input("Enter choice for pop method ")))
        if(f==1):
            lis1.pop()
            print("Last element is removed from your list ")
            print(lis1)
        elif(f==2):
            popind=(int(input("Enter a index for removing element ")))
            lis1.pop(popind)
            print(lis1)
        else:
            print("Invalid choice")
    elif(e==2):
        print(lis1)
        remo1=(eval(input("Enter element name which you want to remove from list ")))
        lis1.remove(remo1)
        print(lis1)
    elif(e==3):
        lis1.clear()
        print("You just cleared the element from your list")
        print(lis1)
    elif(e==4):
        print("Enter 1 for deleting elements using indexing ")
        print("Enter 2 for deleting elements using slicing ")
        print("Enter 3 for deleting the entire list with the memory")
        del1=(int(input("Enter choice for del ")))
        if(del1==1):
            ind=(int(input("Enter index of element you want to delete ")))
            del lis1[ind]
            print(lis1)
        elif(del1==2):
            startsli=(int(input("Enter start index for slicing ")))
            stopsli=(int(input("Enter stop index for slicing ")))
            stepsli=(int(input("Enter steps for slicing ")))
            del lis1[startsli:stopsli:stepsli]
            print(lis1)
        elif(del1==3):
            del lis1
            print("You just deleted the list with the memory")
            print(lis1)
        else:
            print("Invalid choice")
    else:
        print("Invalid choice")
else:
    print("Invalid choice")
        
```
