```python
import random
GamePlay = ["Rock", "Paper", "Scissor"]
player = False
cpu_score = 0
player_score = 0
while True:
    player=input('Press R For Rock\nPress P For Paper\nPress S for Scissor\nPress other to quit : ')
    player=player.capitalize()
    if player == "R":
        if random.choice(GamePlay) == "Paper":
            print('-'*50)
            print("You Lost!!!Better Luck in next match")
            print('-'*50)
            cpu_score+=1
        elif random.choice(GamePlay) == "Scissor":
            print('-'*50)
            print("Congratulation!!! You won!Lets try again")
            print('-'*50)
            player_score+=1
        else:
            print('-'*50)
            print("It's a tie,Let's Play Again!")
            print('-'*50)
    elif player == "P":
        if random.choice(GamePlay) == "Scissor":
            print('-'*50)
            print("You Lost!!!Better Luck in next match")
            print('-'*50)
            cpu_score+=1
        elif random.choice(GamePlay) == "Rock":
            print('-'*50)
            print("Congratulation!!! You won!Lets try again")
            print('-'*50)
            player_score+=1
        else:
            print('-'*50)
            print("It's a tie,Let's Play Again!")
            print('-'*50)
    elif player == "S":
        if random.choice(GamePlay) == "Rock":
            print('-'*50)
            print("You Lost!!!Better Luck in next match")
            print('-'*50)
            cpu_score+=1
        elif random.choice(GamePlay) == "Paper":
            print('-'*50)
            print("Congratulation!!! You won!Lets try again")
            print('-'*50)
            player_score+=1
        else:
            print('-'*50)
            print("It's a tie,Let's Play Again!")
            print('-'*50)
    else:
        print('='*50)
        print("Final Scores:")
        print('-'*50)
        print(f"CPU   :{cpu_score}")
        print(f"Player:{player_score}")
        print('-'*50)
        if cpu_score>player_score:
            print('Hurray! I defeated You!')
            print('='*50)
        elif cpu_score==player_score:
            print('It was a tie contest')
            print('='*50)
        else:
            print('Congratulations! you are the victorious one')
            print('='*50)
        break
```
